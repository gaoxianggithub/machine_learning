## 深度学习是什么：
深度学习是一种计算机技术，通过使用多层神经网络从前一层接收输入并逐步对其进行提取和转换，用于从人类语音识别到动物图像分类等各种用例。这些层级通过最小化误差和提高准确性的算法进行训练。通过这种方式，网络学会执行指定的任务。我们将在下一节详细讨论训练算法。

深度学习具有强大、灵活和简单的特点。这就是为什么我们认为它应该在许多领域中应用的原因。这些领域包括社会科学、自然科学、艺术、医学、金融、科学研究等等。举个个人例子，尽管没有医学背景，Jeremy 创办了一家名为 Enlitic 的公司，该公司使用深度学习算法来诊断疾病。在公司成立几个月后，有消息称其算法能够比放射科医生更准确地识别恶性肿瘤。

## 神经网络：简史
1943年，神经生理学家沃伦·麦卡洛克（Warren McCulloch）和逻辑学家沃尔特·皮茨（Walter Pitts）联手开发了人工神经元的数学模型。在他们的论文“神经活动中内在思想的逻辑演算”中，他们宣称：
> 由于神经活动的“全有或全无”特征，神经事件及其之间的关系可以通过命题逻辑来治疗。发现每个网络的行为都可以用这些术语来描述。
McCulloch和Pitts意识到，可以使用简单的加法和阈值来表示真实神经元的简化模型，如图所示。皮茨是自学成才的，12岁时，他收到了剑桥大学与伟大的伯特兰·罗素（Bertrand Russell）一起学习的邀请。他没有接受这个邀请，事实上，在他的一生中，他没有接受任何高级学位或权威职位的邀请。他的大部分著名工作都是在他无家可归时完成的。尽管他缺乏官方认可的职位，并且社会孤立感日益增加，但他与麦卡洛克的合作很有影响力，并被一位名叫弗兰克·罗森布拉特的心理学家所接受。
![image](https://nbviewer.org/github/fastai/fastbook/blob/master/images/chapter7_neuron.png)

罗森布拉特进一步发展了人工神经元，使其具有学习能力。更重要的是，他致力于制造第一个实际使用这些原理的设备，Mark I Perceptron。在《智能自动机的设计》（The Design of an Intelligent Automaton）一书中，罗森布拉特（Rosenblatt）这样写道：“我们现在即将见证这样一台机器的诞生——一台能够在没有任何人类训练或控制的情况下感知、识别和识别周围环境的机器。感知器建成，能够成功识别简单的形状。

麻省理工学院教授马文·明斯基（Marvin Minsky）（在同一所高中比罗森布拉特落后一个年级！）和西摩·派普特（Seymour Papert）一起写了一本名为《感知器》（麻省理工学院出版社）的书，讲述了罗森布拉特的发明。他们表明，这些设备的单层无法学习一些简单但关键的数学函数（如XOR）。在同一本书中，他们还表明，使用多层设备可以解决这些限制。不幸的是，这些见解中只有第一个得到广泛认可。结果，在接下来的二十年里，全球学术界几乎完全放弃了神经网络。

也许过去50年来神经网络中最关键的工作是David Rumelhart，James McClellan和PDP研究小组的多卷并行分布式处理（PDP），由麻省理工学院出版社于1986年发布。 第1章提出了与罗森布拉特类似的希望：
> 人们比今天的计算机更聪明，因为大脑采用基本的计算架构，更适合处理人们擅长的自然信息处理任务的核心方面。...我们将介绍一个用于建模认知过程的计算框架，该框架似乎...比其他框架更接近计算风格，因为它可能由大脑完成。

PDP在这里使用的前提是，传统的计算机程序的工作方式与大脑非常不同，这可能就是为什么计算机程序（在这一点上）在做大脑认为容易的事情（例如识别图片中的物体）方面如此糟糕的原因。作者声称PDP方法“更接近” 比其他框架“来描述大脑的工作方式，因此它可能能够更好地处理这些类型的任务。

事实上，PDP 中列出的方法与当今神经网络中使用的方法非常相似。该书将并行分布式处理定义为需要：

1. 一组处理单元
2. 激活状态
3. 每个单元的输出功能
4. 单元之间的连接模式
5. 通过连接网络传播活动模式的传播规则
6. 一种激活规则，用于将冲击单元的输入与该单元的当前状态相结合，以生成单元的输出
7. 一种学习规则，其中连接模式由经验修改
8. 系统必须在其中运行的环境
我们将在本书中看到现代神经网络处理这些需求中的每一个。

在1980年代，大多数模型都是用第二层神经元构建的，从而避免了Minsky和Papert发现的问题（这是他们的“单元之间的连接模式”，使用上面的框架）。事实上，神经网络在 80 年代和 90 年代被广泛用于真正的、实用的项目。然而，对理论问题的误解再次阻碍了该领域。从理论上讲，只需增加一层额外的神经元就足以使任何数学函数与这些神经网络近似，但在实践中，这种网络通常太大，太慢而无法使用。

尽管研究人员在30年前就表明，要获得实用的良好性能，您需要使用更多的神经元层，但直到最近十年，这一原理才得到更广泛的赞赏和应用。神经网络现在终于发挥了它们的潜力，这要归功于使用更多的层，再加上由于计算机硬件的改进、数据可用性的提高以及算法调整，使神经网络能够更快、更容易地训练。我们现在有了罗森布拉特所承诺的：“一台能够在没有任何人类训练或控制的情况下感知、识别和识别周围环境的机器。

````python
from fastai.vision.all import *
# 建议用浏览器下载这个链接，程序下载太慢
# print(URLs.PETS)

path = untar_data(URLs.PETS)/'images'

def is_cat(x): 
    return x[0].isupper()

dls = ImageDataLoaders.from_name_func(
    path, get_image_files(path), valid_pct=0.2, seed=42,
    label_func=is_cat, item_tfms=Resize(224))
learn = vision_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(1)
````

### 什么是机器学习？
分类器是深度学习模型。如前所述，深度学习模型使用神经网络，该网络最初可以追溯到 1950 年代，并且由于最近的进步而最近变得强大。

另一个关键的背景是，深度学习只是机器学习更一般学科中的一个现代领域。要了解您在训练自己的分类模型时所做的事情的本质，您不需要了解深度学习。了解您的模型和训练过程如何成为适用于一般机器学习的概念示例就足够了。

因此，在本节中，我们将描述什么是机器学习。我们将研究关键概念，并展示如何将它们追溯到介绍它们的原始文章。

机器学习与常规编程一样，是一种让计算机完成特定任务的方法。但是，我们将如何使用常规编程来完成我们在上一节中所做的工作：识别照片中的狗与猫？我们必须为计算机写下完成任务所需的确切步骤。

通常，当我们编写程序时，我们很容易写下完成任务的步骤。我们只是考虑如果我们必须手动完成任务，我们会采取的步骤，然后将它们转换为代码。例如，我们可以编写一个对列表进行排序的函数。通常，我们会编写一个类似于下图的函数（其中输入可能是未排序的列表，结果是排序列表）。

![image](https://gitee.com/shawn2gao/myimg/raw/master/2023-5-1516:53:55.png1684140834775.png)

但是对于识别照片中的物体来说，这有点棘手;当我们识别图片中的物体时，我们采取哪些步骤？我们真的不知道，因为这一切都发生在我们的大脑中，而我们却没有意识到它！

早在1949年，在计算的黎明时期，一位名叫亚瑟·塞缪尔（Arthur Samuel）的IBM研究人员开始研究一种不同的方法，让计算机完成任务，他称之为机器学习。在他1962年的经典文章“人工智能：自动化的前沿”中，他写道：

> 为这种计算编写计算机充其量是一项艰巨的任务，主要不是因为计算机本身固有的复杂性，而是因为需要以最令人恼火的细节阐明该过程的每一分钟步骤。正如任何程序员都会告诉你的那样，计算机是巨大的白痴，而不是巨大的大脑。

他的基本想法是：与其告诉计算机解决问题所需的确切步骤，不如向它展示要解决的问题的示例，并让它自己弄清楚如何解决它。事实证明，这非常有效：到 1961 年，他的跳棋程序已经学到了很多东西，以至于它击败了康涅狄格州冠军！以下是他如何描述他的想法（来自与上面的同一篇文章）：

> ：假设我们安排了一些自动方法来测试任何当前权重分配在实际性能方面的有效性，并提供一种改变权重分配的机制，以最大限度地提高性能。我们不需要详细讨论这样一个程序，就能看到它可以完全自动化，并且看到这样编程的机器会从它的经验中“学习”。

这个简短的陈述中嵌入了许多强大的概念：

* “权重分配”的想法
* 事实上，每个权重分配都有一些“实际性能”
* 要求有一种“自动手段”来测试该性能，
* 需要一个“机制”（即另一个自动过程）来通过改变权重分配来提高性能
让我们一一了解这些概念，以便了解它们在实践中如何组合在一起。首先，我们需要理解塞缪尔所说的权重分配是什么意思。



权重只是变量，权重分配是这些变量值的特定选择。程序的输入是它为了产生结果而处理的值，例如，将图像像素作为输入，并返回分类“狗”。程序的权重分配是定义程序运行方式的其他值。

由于它们会影响程序，因此它们在某种意义上是另一种输入，因此我们将在中更新我们的基本图片并将其替换为以考虑到这一点。
![image](https://gitee.com/shawn2gao/myimg/raw/master/2023-5-1516:58:49.png1684141128340.png)

我们已将盒子的名称从程序更改为模型。这是为了遵循现代术语，并反映模型是一种特殊的程序：它是一个可以做很多不同事情的程序，具体取决于权重。它可以通过许多不同的方式实现。例如，在Samuel的跳棋程序中，不同的权重值会导致不同的跳棋策略。

（顺便说一下，塞缪尔所说的“权重”现在通常被称为模型参数，以防您遇到该术语。术语权重是为特定类型的模型参数保留的。

接下来，塞缪尔说，我们需要一种自动方法来测试当前任何重量分配在实际性能方面的有效性。在他的跳棋程序的情况下，模型的“实际性能”将是它的播放效果。您可以通过将两个模型设置为相互对战并查看哪个模型通常获胜来自动测试它们的性能。

最后，他说我们需要一种改变权重分配的机制，以最大限度地提高性能。例如，我们可以查看获胜模型和失败模型之间的权重差异，并在获胜方向上进一步调整权重。

我们现在可以看到为什么他说这样的程序可以完全自动化，并且......如此编程的机器将从其经验中“学习”。当权重的调整也是自动的时，学习将变得完全自动 - 当我们不是通过手动调整权重来改进模型时，我们依赖于基于性能进行调整的自动化机制。

### 什么是神经网络？
不难想象跳棋程序的模型会是什么样子。可能会有一系列编码的跳棋策略和某种搜索机制，然后权重可能会改变策略的选择方式，在搜索过程中关注棋盘的哪些部分，等等。但是，对于图像识别程序，理解文本或我们可能想象的许多其他有趣问题，模型的外观并不明显。

> 注意：分类和回归：分类和回归在机器学习中具有非常具体的含义。这是我们将在本书中研究的两种主要模型类型。分类模型是尝试预测类或类别的模型。也就是说，它从许多离散的可能性进行预测，例如“狗”或“猫”。回归模型是尝试预测一个或多个数字量（例如温度或位置）的模型。有时人们使用回归一词来指代一种称为线性回归模型的特定类型的模型;这是一个不好的做法，我们不会在本书中使用该术语！

这里要提到的最重要的参数是valid_pct=0.2。这告诉fastai保留20%的数据，并且在训练模型时不使用它们。这20%的数据被称为验证集；剩下的80%被称为训练集。验证集用于衡量模型的准确性。默认情况下，被保留的20%是随机选择的。参数seed=42将随机种子设置为相同的值，这样每次运行这段代码时都会得到相同的验证集。这样，如果我们改变模型并重新训练它，我们知道任何差异都是由于模型的变化，而不是由于具有不同的随机验证集。

fastai始终只使用验证集来显示模型的准确性，而不使用训练集。这是非常关键的，因为如果你训练一个足够大、时间足够长的模型，它最终将记住数据集中每个项的标签！这样得到的结果实际上不是一个有用的模型，因为我们关心的是模型在以前未见过的图片上的表现。这总是我们创建模型的目标：使其对模型仅在训练后未来才能看到的数据有用。

即使你的模型没有完全记住所有的数据，在训练的早期阶段，它可能已经记住了其中的某些部分。因此，你训练的时间越长，训练集的准确性就会提高；验证集的准确性也会一段时间内提高，但最终随着模型开始记住训练集而不是发现数据中可推广的潜在模式，验证集的准确性会开始变差。当发生这种情况时，我们说模型过拟合了。

在训练所有机器学习从业者和所有算法时，**过拟合**是最重要和最具挑战性的问题。如您所见，创建一个模型非常容易，该模型在对已训练的确切数据进行预测方面做得很好，但是对模型以前从未见过的数据进行准确预测要困难得多。当然，这是在实践中真正重要的数据。例如，如果你创建一个手写数字分类器（我们很快就会！）并使用它来识别写在支票上的数字，那么你永远不会看到模型训练的任何数字 - 支票将有略有不同的书写变化来处理。在本书中，您将学习许多避免过度拟合的方法。但是，只有在确认确实发生了过拟合之后，才应使用这些方法（即，您实际上观察到验证准确性在训练期间越来越差）。我们经常看到从业者使用过度拟合避免技术，即使他们有足够的数据不需要这样做，最终得到的模型可能不如他们所能达到的准确。

在fastai中有许多不同的架构，我们将在本书中介绍它们（以及讨论如何创建自己的架构）。然而，在深度学习过程中，选择一个架构通常并不是非常重要的部分。这是学术界喜欢讨论的话题，但在实践中，你不太可能需要花费太多时间在这上面。有一些标准的架构在大多数情况下都有效，而在这种情况下，我们使用的是一种称为ResNet的架构，在本书中我们将对其进行大量讨论；它对许多数据集和问题都具有快速和准确的特点。resnet34中的34指的是这种架构变种中的层数（其他选项有18、50、101和152）。使用更多层的架构的模型训练时间更长，更容易过拟合（即在验证集上的准确性开始变差之前无法训练它们的周期数更少）。另一方面，当使用更多的数据时，它们可以更准确一些。

什么是指标？指标是使用验证集衡量模型预测质量的函数，并将在每个纪元结束时打印。在本例中，我们使用`error_rate` ，这是 fastai 提供的一个函数，它所做的就是它所说的：告诉您验证集中未正确分类的图像百分比。分类的另一个常见指标是准确率（`1 - error_rate`）。Fastai提供了更多内容，将在本书中讨论。

指标的概念可能会让您想起损失，但有一个重要的区别。损失的全部目的是定义训练系统可以用来自动更新权重的“性能度量”。换句话说，损失的好选择是随机梯度下降易于使用的选择。但是指标是为人类消费而定义的，因此一个好的指标是易于理解的指标，并且尽可能接近您希望模型执行的操作。有时，您可能会认为损失函数是一个合适的指标，但事实并非如此。

vision_learner还有一个参数，默认为（因此在这种情况下使用它，即使我们尚未指定它），它将模型中的权重设置为已由专家训练的值，以识别 1 万张照片中的一千个不同类别（使用著名的 ImageNet 数据集）。具有已在其他数据集上训练的权重的模型称为预训练模型。您几乎应该始终使用预训练模型，因为这意味着您的模型在向它展示任何数据之前就已经非常强大了。而且，正如你将看到的，在深度学习模型中，这些功能中的许多都是你需要的，几乎与项目的细节无关。例如，预训练模型的一部分将处理边缘、渐变和颜色检测，这是许多任务所必需的。

使用预训练模型时， 将删除最后一层，因为它始终专门针对原始训练任务（即 ImageNet 数据集分类）进行定制，并将其替换为一个或多个具有随机权重的新层，这些新层的大小适合您正在使用的数据集。模型的最后一部分称为头部。vision_learner

使用预训练模型是我们必须的最重要的方法，使我们能够更快地训练更准确的模型，使用更少的数据，更少的时间和金钱。你可能会认为这意味着使用预训练模型将是学术深度学习中研究最多的领域......但你会非常非常错误！预训练模型的重要性在大多数课程、书籍或软件库功能中通常不被认可或讨论，在学术论文中也很少被考虑。当我们在2020年初写这篇文章时，事情才刚刚开始发生变化，但可能需要一段时间。所以要小心：与你交谈的大多数人可能会大大低估你在深度学习中用很少的资源所能做的事情，因为他们可能不会深入了解如何使用预训练模型。

将预训练模型用于与最初训练不同的任务称为迁移学习。不幸的是，由于迁移学习的研究不足，很少有领域有可用的预训练模型。例如，目前医学中可用的预训练模型很少，这使得迁移学习在该领域使用具有挑战性。此外，目前还不太了解如何将迁移学习用于时间序列分析等任务。
> 行话：迁移学习：将预训练模型用于与最初训练不同,但指标相似的任务。

我们代码的第六行告诉fastai如何拟合模型：
`learn.fine_tune(1)`
正如我们所讨论的，该架构仅描述了数学函数的模板;在我们为其包含的数百万个参数提供值之前，它实际上不会执行任何操作。

这是深度学习的关键——确定如何拟合模型的参数以使其解决您的问题。为了拟合模型，我们必须提供至少一条信息：查看每个图像的次数（称为 epoch 数）。您选择的纪元数在很大程度上取决于您有多少可用时间，以及您发现在实践中拟合模型所需的时间。如果选择的数字太小，则以后始终可以训练更多纪元。

但是为什么这种方法叫fine_tune，而不是fit? Fastai实际上确实有一个名为fit，它确实适合模型（即多次查看训练集中的图像，每次更新参数以使预测越来越接近目标标签）。但在这种情况下，我们从一个预训练模型开始，我们不想丢弃它已经拥有的所有功能。正如你将在本书中学到的，有一些重要的技巧可以使预训练模型适应新的数据集——这个过程称为`fine-tuning`。

行话： fine_tuning：一种迁移学习技术，其中预训练模型的参数通过使用与用于预训练的任务不同的任务来训练额外的 epoch。

当您使用fine_tune方法时，fastai 将为您使用这些技巧。您可以设置一些参数（我们将在后面讨论），但在此处显示的默认形式中，它执行两个步骤：

1. 使用一个纪元来拟合模型的那些部分，以使新的随机头正确处理数据集。
2. 使用调用方法时请求的 epoch 数来拟合整个模型，更新后面层（尤其是头部）的权重比前面的层快（正如我们将看到的，通常不需要对预训练的权重进行太多更改）。
   
模型的头部是新添加的特定于新数据集的部分。纪元是一次完整的数据集传递。调用fit后，将打印每个纪元后的结果，显示纪元编号、训练和验证集损失（用于训练模型的“性能度量”）以及您请求的任何指标（在本例中为错误率）。

因此，通过所有这些代码，我们的模型学会了仅从标记的例子中识别猫和狗。但它是如何做到的呢？

## 术语总结
| 术语 | 含义|
|-|-|
|标签(Label) | 我们试图预测的数据，比如“狗”或“猫”|
|架构(Architecture) | 我们试图适应的模型的模板；实际上是将输入数据和参数传递给的数学函数|
|模型(Model) | 架构与特定参数集的组合|
|参数(Parameters) | 模型中的值，可以改变模型能够执行的任务，并通过模型训练进行更新|
|适应(Fit) | 更新模型的参数，使得使用输入数据的模型预测与目标标签匹配|
|训练(Train) | 与“适应”同义词|
|预训练模型(Pretrained model) | 已经通过使用大型数据集进行训练的模型，并将其进行微调|
|微调(Fine-tune) | 为不同的任务更新预训练模型|
|时期(Epoch) | 对输入数据的完整遍历|
|损失(Loss) | 衡量模型好坏的指标，通过 SGD 选择以推动训练|
|度量(Metric) | 使用验证集衡量模型好坏的测量指标，为人类消费选择|
|验证集(Validation set) | 从训练中排除的一组数据，仅用于衡量模型好坏|
|训练集(Training set) | 用于适应模型的数据；不包括验证集中的任何数据|
|过拟合(Overfitting) | 以一种记住输入数据特定特征的方式训练模型，而不是很好地推广到训练期间未见的数据|
|CNN | 卷积神经网络；一种特别适用于计算机视觉任务的神经网络类型|

有了这些词汇，我们现在就可以将迄今为止介绍的所有关键概念汇集在一起。花点时间查看这些定义并阅读以下摘要。如果你能遵循解释，那么你就有能力理解即将到来的讨论。

机器学习是一门学科，我们定义一个程序不是完全通过自己编写，而是通过从数据中学习。深度学习是机器学习中的一个专业，它使用多层神经网络。图像分类是一个代表性的例子（也称为图像识别）。我们从标记的数据开始;也就是说，一组图像，我们为每个图像分配了一个标签，指示它所代表的内容。我们的目标是生成一个称为模型的程序，给定一个新图像，它将对新图像所代表的内容做出准确的预测。

每个模型都从架构的选择开始，这是该模型如何在内部工作的通用模板。训练（或拟合）模型的过程是查找一组参数值（或权重）的过程，这些参数值（或权重）将该通用体系结构专用于适用于我们特定类型数据的模型。为了定义模型在单个预测上的表现，我们需要定义一个损失函数，它决定了我们如何对预测进行好坏评分。

为了使训练过程更快，我们可以从一个预训练模型开始，一个已经在其他人的数据上训练的模型。然后，我们可以通过在数据上进行更多训练来使其适应我们的数据，这个过程称为微调。

当我们训练一个模型时，一个关键问题是确保我们的模型泛化，也就是说，它从我们的数据中学习一般教训，这些教训也适用于它将遇到的新项目，以便它可以对这些项目做出良好的预测。风险在于，如果我们对模型的训练不好，而不是学习一般课程，它有效地记住了它已经看到的东西，然后它会对新图像做出糟糕的预测。这种故障称为过度拟合。为了避免这种情况，我们总是将数据分为两部分，训练集和验证集。我们通过仅显示训练集来训练模型，然后通过查看模型在验证集中的项目上的表现来评估模型的表现。通过这种方式，我们检查模型从训练集中学到的教训是否是推广到验证集的教训。为了让一个人评估模型在验证集上的整体表现如何，我们定义了一个指标。在训练过程中，当模型看到训练集中的每个项目时，我们称之为纪元。

所有这些概念都适用于一般的机器学习。也就是说，它们适用于各种方案，通过用数据训练模型来定义模型。使深度学习与众不同的是一类特殊的架构：基于神经网络的架构。特别是，像图像分类这样的任务严重依赖于卷积神经网络，我们将很快讨论。

> 行话：Tabular：表格形式的数据，例如来自电子表格、数据库或 CSV 文件的数据。表格模型是一种模型，它尝试根据表的其他列中的信息预测表的一列。

事实证明，看起来也非常相似。以下是训练模型所需的代码，该模型将根据其社会经济背景预测一个人是否是高收入者：

````python
from fastai.tabular.all import *
path = untar_data(URLs.ADULT_SAMPLE)

dls = TabularDataLoaders.from_csv(path/'adult.csv', path=path, y_names="salary",
    cat_names = ['workclass', 'education', 'marital-status', 'occupation',
                 'relationship', 'race'],
    cont_names = ['age', 'fnlwgt', 'education-num'],
    procs = [Categorify, FillMissing, Normalize])

learn = tabular_learner(dls, metrics=accuracy)
````


正如你所看到的，我们需要告诉fastai哪些列是分类变量（即包含一组离散选择的值，例如职业），哪些是连续变量（即包含表示数量的数字，例如年龄）。

针对这个任务，没有可用的预训练模型（一般来说，预训练模型在任何表格建模任务中并不广泛存在，尽管有些组织为内部使用创建了这样的模型），所以我们在这种情况下不使用fine_tune。相反，我们使用fit_one_cycle，这是训练fastai模型的最常用方法，可以从头开始训练（即不使用迁移学习）。

````python
learn.fit_one_cycle(3)
````
